```python
import os
import cv2
from matplotlib import pyplot as plt
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import load_model
# Load the trained model
model_path = '/content/drive/MyDrive/model/Model.h5'#path to the model file 
model = load_model(model_path)

def test_image(path):
    # Read the image
    img = cv2.imread(path)

    # Resize the image
    resize = tf.image.resize(img, (256, 256))

    # Display the resized image
    plt.imshow(resize.numpy().astype(int))
    plt.show()

    # Preprocess the image and make prediction
    result = model.predict(tf.expand_dims(resize / 255, 0))

    if result > 0.5:
        print('The above image is Real.')
    else:
        print('The above image is Fake.')

# Get user input for image path
image_path = input('Enter the path of the image to test: ')

# Call the test_image function with the provided image path
test_image(image_path)
```
